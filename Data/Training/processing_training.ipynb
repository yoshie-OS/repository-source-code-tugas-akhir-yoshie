```python
import pandas as pd
```

```python
train_df = pd.read_csv("combined_emotion_dataset.csv")
```

```python
print(train_df['label'].unique())
```

```python
label_map = {
    'happy': 'joy',
    'Joy': 'joy',
    'anger': 'anger',
    'Anger': 'anger',
    'fear': 'fear',
    'Fear': 'fear',
    'sadness': 'sadness',
    'Sad': 'sadness',
    'love': 'love',
    'Love': 'love',
}

train_df['label'] = train_df['label'].map(label_map)

print(train_df['label'].value_counts())
print(train_df['label'].isna().sum())  # pastiin ga ada yang ke-miss
```

```python
train_df.to_csv("training_dataset.csv", index=False)
print("Saved.")
```
